```python


df = pd.DataFrame()
for person in data:
    if len(person['GAMELOG']) == 0:
        continue
    #only takes data from those dict keys as columns, no need to drop unnecessary
    df1 = pd.DataFrame(person, columns=['DISPLAY_FIRST_LAST', 'PERSON_ID', 'TEAM_ID'], index=[0])
    df2 = pd.DataFrame(person['GAMELOG'])
    merged = pd.merge(df1, df2, left_on='PERSON_ID', right_on='Player_ID', how='outer')
    df = pd.concat([df, merged])
    
#each merge would end up repeated the index from each merge table, 
#because Pd does not require unique indeces, so drop, and then reset to inplace
df.reset_index(drop=True, inplace=True)

```

## 3:
How many players played in at least 82 games last season?

```python
def at_least_82(df):
    #initialize list of players
    players_over_82 = []
    grouped_players = group_players(df)
    for player in grouped_players:
        player_name = player[0]
        num_of_games = len(player[1])
        if num_of_games >= 82:
            players_over_82.append(player_name)
    return players_over_82
```

## 7:

Find the player who scored the least amount of points while playing in at least 40 games last season.

```python
def filter_40_plus_games(df):
    players_grouped = group_players(df)
    filtered_players_40 = players_grouped.filter(lambda x: len(x) > 40)
    return filtered_players_40

def least_points_with_40_games(df):
    filtered_players = filter_40_plus_games(df)
    sum_points = filtered_players.groupby("DISPLAY_FIRST_LAST")['PTS'].sum()
    least_scorer = sum_points.nsmallest(1)
    return least_scorer
```

## 8: 

What is the probability that a random player had a game where they scored more than 40 points?

```python
#retrieve total number of unique players
def total_players(df):
    grouped_players = group_players(df)
    return len(grouped_players)

#calculate odds of getting num1 in sample space num2
def calculate_odds(num1, num2):
    return (num1/num2)
```

## 9:

What is the probability that a randomly selected player from last season would average more than 20 points per game?

```python
import statistics

#return a DF of all the games featuring players who have scored over x amount
def attribute_over_x(df, attribute, x):
    grouped_players = group_players(df)
    #filter by selecting where mean of attribute is greater than amount
    over_x_avg = grouped_players.filter(lambda y : statistics.mean(y[attribute]) > x)
    return over_x_avg

def num_of_20_ppg(df):
    #get data_frame of all games with players who have 20 or better average
    players_over_20 = attribute_over_x(df, 'PTS', 20)
    grouped_players = group_players(players_over_20)
    #regroup and get the length to get count
    return len(grouped_players)


#probability of getting a player with 20 plus ppg average at random
def probs_of_20_avg(df):
    players_total = total_players(df)
    over_20_ppg_players = num_of_20_ppg(df)
    return calculate_odds(over_20_ppg_players, players_total)
```

## 12:

What is the probability that a randomly selected player had at least one game in which they scored 40 points last season. 

```python

def scored_40_plus(df, attribute):
    grouped_players = group_players(df)
    #filter by selecting where max of the attribute is over 40
    attr_over_40 = grouped_players.filter(lambda x : x[attribute].max() > 40)
    return attr_over_40

def num_of_40_plus(df):
    players_over_40_pts = scored_40_plus(df, 'PTS')
    grouped_over_40 = group_players(players_over_40_pts)
    num_over_40 = len(grouped_over_40)
    return num_over_40

def odds_of_40_pt_game(df):
    total = total_players(df)
    num_40_pt_players = num_of_40_plus(df)
    return calculate_odds(num_40_pt_players, total)
```
